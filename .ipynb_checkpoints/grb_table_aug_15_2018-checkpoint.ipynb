# Numeric Analysis

### PCA

```python
fermi_numeric = fermi_clean[['name', 'trigger_algorithm', 'reliability', 'channel_low', 'channel_high', 'trigger_timescale', 'adc_low', 'adc_high']]
```

```python
fermi_numeric.head()
```

```python
# Standardize

from sklearn.preprocessing import StandardScaler
features = ['channel_low', 'adc_low', 'channel_high', 'adc_high']
# Separating out the features
x = fermi_numeric.loc[:, features].values
# Separating out the target
y = fermi_numeric.loc[:,['name']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
# PCA 

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, fermi_numeric[['trigger_algorithm']]], axis = 1)
finalDf.head(20)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = [119, 117, 16]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['trigger_algorithm'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

## Similarity

```python
# Jaccard Similarity score

from sklearn.metrics import jaccard_similarity_score
print(jaccard_similarity_score(fermi_numeric['channel_low'], fermi_numeric['adc_low']))
```

```python
# Hamming Distance

from sklearn.metrics.pairwise import pairwise_distances
print(1 - pairwise_distances(fermi_numeric.T, metric = "hamming"))
```
